# Business Understanding

Sebuah perusahaan e-commerce "X" memiliki masalah berupa turunnya jumlah transaksi pelanggan dalam beberapa waktu terakhir. Rendahnya jumlah transaksi berujung pada menurunnya jumlah pendapatan perusahaan sehingga menjadi sinyal berbahaya untuk pertumbuhan perusahaan kedepannya. Setelah dilakukan pengusutan lebih lanjut, diketahui bahwa terdapat penurunan yang signifikan dalam jumlah konversi kunjungan pengguna website yang berujung pada transaksi dalam beberapa waktu terakhir. Beberapa faktor diduga menjadi penyebabnya, akan tetapi perusahaan belum bisa mengambil kesimpulan seutuhnya. Perusahaan perlu melakukan analisis terkait faktor-faktor yang diduga mempengaruhi keputusan pengunjung melakukan transaksi/tidak. Oleh karena itu, dengan melihat data historis, perusahaan meminta anda untuk membuat model klasifikasi apakah suatu session akan berujung pada transaksi/tidak berdasarkan informasi pada session tersebut.

# Data Understanding


Dataset berisi informasi tentang setiap session di sebuah website e-commerce X. Session merupakan peristiwa seseorang mengunjungi website e-commerce X. Session dimulai ketika orang tersebut mengunjungi website e-commerce X di salahsatu tab browsernya, hingga mereka menutup tab tersebut/mengunjungi website lain pada tab tersebut. Informasi tentang satu session dimuat dalam satu baris data.
Informasi yang dimuat terdiri atas:

1.	**Jenis Page yang dikunjungi**

    Terdapat 3 jenis page pada website, yakni:

    1)	Administrative

    2)	Informational

    3)	Product Related

    Keterangan jumlah page yang dikunjungi untuk setiap jenis page dinyatakan dalam kolom-kolom (JenisPage). Sebagai contoh : Jumlah page berjenis administrative yg dikunjungi dalam suatu session dinyatakan dalam kolom “Administrative”.
    
    Sedangkan, total durasi suatu jenis page dikunjungi dalam suatu session dinyatakan dalam kolom (JenisPage Duration). Sebagai contoh: Total durasi page berjenis administrative dikunjungi pada suatu session dinyatakan dalam kolom “Administrative Duration”.


2.	**Bounce Rate**

    Bounce rate adalah persentase pengunjung yang langsung meninggalkan website setelah membuka satu halaman saja. Parameter ini dimiliki oleh setiap halaman dalam website. Misalnya, bounce rate untuk halaman daftar produk makanan adalah persentase pengunjung yang masuk ke website ecommerce melalui halaman daftar produk makanan, kemudian langsung keluar dari website setelahnya (tidak membuka halaman-halaman lain di website).
    
    Pada dataset ini, nilai bounce rate untuk setiap row (session) merupakan rata-rata dari bounce rate setiap halaman yang dikunjungi pada session tersebut selama periode tertentu.


3.	**Exit Rate**

    Exit rate adalah persentase pengunjung yang meninggalkan website setelah mendatangi suatu halaman. Parameter ini dimiliki oleh setiap halaman dalam website. Misalnya, exit rate untuk halaman daftar produk makanan adalah persentase pengunjung yang keluar dari website ecommerce setelah mendatangi halaman daftar produk makanan.
    
    Pada dataset ini, nilai exit rate untuk setiap row (session) merupakan rata-rata dari exit rate setiap halaman yang dikunjungi pada session tersebut selama periode tertentu.


4.	**Page Value**

    Page value adalah nilai dari suatu halaman pada website yang dihitung berdasarkan rata-rata jumlah transaksi yang dilakukan oleh pengunjung setelah terlebih dahulu mengunjungi halaman tersebut.

    Sebagai contoh, dalam suatu session, pengunjung melakukan transaksi sebesar 1 juta rupiah setelah melewati berbagai halaman termasuk halaman daftar produk makanan. Maka, page value dari halaman daftar produk makanan menjadi 1 juta rupiah. Kemudian, apabila pada kunjungan ke halaman daftar produk makanan pada session berikutnya pengunjung tidak melakukan transaksi, maka page value menjadi 500.000 rupiah.

    Pada dataset ini, page value untuk setiap row (session) merupakan rata-rata dari page value setiap halaman yang dikunjungi pada session tersebut selama periode tertentu.

5.	**Month**

    Menyatakan bulan kalender ketika suatu session itu terjadi

6.	**Special Day**

    Menyatakan ukuran kedekatan waktu session terjadi dengan hari special (seperti valentine, lebaran, dan sejenisnya). Ukuran kedekatan dinyatakan dalam rentang 0-1. Nilai 1 diberikan apabila jarak antara waktu session dengan hari special kurang dari 2 hari, sedangkan nilai 0 diberikan apabila jaraknya lebih dari 10 hari.

7.	**Operating Systems, Browser, Region, Traffic Type**

    4 informasi ini bersifat variabel kategorikal. Operating systems menyatakan jenis operating systems (yang dinotasikan sebagai angka 1-4; misal 1 = Windows, 2= Apple dst.) yang digunakan oleh pengunjung pada suatu session. Misalnya, untuk session pertama, jenis operating systems yang digunakan adalah jenis 1 (Windows).

8.	**VisitorType**

    Menyatakan jenis pengunjung website apakah pengunjung baru (Returning_Visitor) atau pengunjung lama yang kembali lagi (New_Visitor)

9.	**Weekend**

    Memberi informasi apakah session terjadi pada weekend (TRUE) atau weekday (FALSE)

10.	**Revenue**

    Informasi yang merupakan variabel output. Menyatakan apakah session berakhir menjadi transaksi (TRUE) atau tidak (FALSE)

Sebelum melaukan pemodelan, kita perlu memisahkan kolom yang merupakan variabel dependen dan independen, di kasus ini, variabel dependen adalah Revenue dan variabel independen adalah variabel lain selain revenue.

**20. Memisah variabel X dan variabel y pada dataset**

```python
# Memisahkan variabel dependen dengan variabel independen
# dengan cara menghilangkan kolom Revenue dengan menggunakan metode drop.
# variabel independen:
x = data.drop(['Revenue'], axis=1)

# variabel dependen:
y = data['Revenue']

# Untuk memastikan bahwa x dan y sudah sesuai yang kita inginkan,
# periksa shape dari x dan y
print("Shape of x: ", x.shape)
print("Shape of y: ", y.shape)
```

```python
# membuat feature names dari data x (untuk decision tree)
feature_names=list(x.columns)
feature_names
```

Sebelum melakukan pemodelan, kita perlu membagi data menjadi dua bagian, yaitu train data, yang digunakan untuk membangun model, dan test data, yang digunakan untuk mengevaluasi model.

**21. Men-split data menjadi training data dan test data**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)

# periksa shape

print("Shape of x_train :", x_train.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of x_test :", x_test.shape)
print("Shape of y_test :", y_test.shape)
```

# Modelling

Pemodelan di kasus ini menggunakan random forest classifier. Random forest classifier ini digunakan untuk memprediksi apakah pengguna akan 'membeli' atau tidak. Random forest classifier adalah salah satu algoritma yang digunakan dalam klasifikasi data dalam jumlah besar.

**22. Random Forest**

Metode random forest merupakan salah satu metode paling umum digunakan dalam task klasifikasi. Pada dasarnya, metode ini adalah metode decision tree namun dilakukan banyak decision tree secara random (maka dari itu dinamakan random forest) yang mana output klasifikasi didasarkan pada vote terbanyak dari decision tree - decision tree yang terbentuk.

## Evaluasi Model

**25. Confusion Matrix**

Confusion matrix juga sering disebut error matrix. Pada dasarnya confusion matrix memberikan informasi perbandingan hasil klasifikasi yang dilakukan oleh sistem (model) dengan hasil klasifikasi sebenarnya. Terdapat beberapa istilah yang digunakan dalam confusion matrix ini
1. True Positive (TP): data positif/yes/1 yang diprediksi benar
2. True Negative (TN): data negative /no/0 terprediksi benar
3. False Positive (FP): data positif/yes/1 yang diprediksi salah. Merupakan Error Tipe I
4. False Negative (FN): data negatif/yes/0 yang diprediksi salah. Merupakan Error Tipe II

```python
# Confusion matrix

from sklearn.metrics import confusion_matrix

# membuat kanvas
fig = plt.figure(figsize=(7,18))
# membagi kanvas menjadi 3 bagian karena kita ingin membandingkan tiga model
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)

# ambil data
# untuk model Random Forest
cm_rf = confusion_matrix(y_test, y_pred_rf) # (y test, y hasil prediksi)
# untuk model Regresi Logistik
cm_lr = confusion_matrix(y_test, y_pred_lr)
# untuk model Decision Tree
cm_dt = confusion_matrix(y_test, y_pred_dt)

# visualisasi heatmap
# untuk model Random Forest
sns.heatmap(cm_rf, ax=ax1, annot = True, fmt='.0f')
# untuk model Regresi Logistik
sns.heatmap(cm_lr, ax=ax2, annot = True, fmt='.0f')
# untuk model Decision Tree
sns.heatmap(cm_dt, ax=ax3, annot = True, fmt='.0f')

# memberi judul
ax1.set_title("Confusion Matrix dari Random Forest", fontdict={'fontsize': '14'})
ax2.set_title("Confusion Matrix dari  Regresi Logistik", fontdict={'fontsize': '14'})
ax3.set_title("Confusion Matrix dari Decision Tree", fontdict={'fontsize': '14'})
ax1.set_xlabel("True Value", fontdict={'fontsize': '12'})
ax1.set_ylabel("Predicted Value", fontdict={'fontsize': '12'})
ax2.set_xlabel("True Value", fontdict={'fontsize': '12'})
ax2.set_ylabel("Predicted Value", fontdict={'fontsize': '12'})
ax3.set_xlabel("True Value", fontdict={'fontsize': '12'})
ax3.set_ylabel("Predicted Value", fontdict={'fontsize': '12'})

plt.show()
```

**26. Accuracy**

Accuracy menggambarkan seberapa akurat model dapat mengklasifikasikan dengan benar. Maka, accuracy merupakan rasio prediksi benar (positif dan negatif) dengan keseluruhan data. Sehingga rumusnya adalah TP+TF/(TP+TF+FP+FN)

```python
from sklearn.metrics import accuracy_score

# accuracy: (tp + tn) / (p + n)
# Random Forest
accuracy_rf = accuracy_score(y_test, y_pred_rf)
# Regresi Logistik
accuracy_lr = accuracy_score(y_test, y_pred_lr)
# Decision Tree
accuracy_dt = accuracy_score(y_test, y_pred_dt)

print('Accuracy Random Forest: ', accuracy_rf)
print('Accuracy Regresi Logistik: ', accuracy_lr)
print('Accuracy Decision Tree: ', accuracy_dt)
```

**27. Precision**
Precision = Type 1 Error / Total Predicted Positive

Precision menggambarkan tingkat keakuratan antara data yang diminta dengan hasil prediksi yang diberikan oleh model. Maka, precision merupakan rasio prediksi benar positif dibandingkan dengan keseluruhan hasil yang diprediksi positf. Sehingga rumusnya adalah TP/(TP+FP).

```python
from sklearn.metrics import precision_score

# precision: tp / (tp + fp)
# Random Forest
precision_rf = precision_score(y_test, y_pred_rf)
# Regresi Logistik
precision_lr = precision_score(y_test, y_pred_lr)
# Decision Tree
precision_dt = precision_score(y_test, y_pred_dt)

print('Precision Random Forest: ', precision_rf)
print('Precision Regresi Logistik: ', precision_lr)
print('Precision Decision Tree: ', precision_dt)
```

**28. Recall**
Recall = Type 2 Error / Total Actual Positive

Recall menggambarkan keberhasilan model dalam menemukan kembali sebuah informasi. Maka, recall merupakan rasio prediksi benar positif dibandingkan dengan keseluruhan data yang benar positif. Sehinnga rumusnya adalah TP / (TP+FN)

```python
from sklearn.metrics import recall_score

# recall: tp / (tp + fn)
# Random Forest
recall_rf = recall_score(y_test, y_pred_rf)
# Regresi Logistik
recall_lr = recall_score(y_test, y_pred_lr)
# Decision Tree
recall_dt = recall_score(y_test, y_pred_dt)

print('Recall Random Forest: ', recall_rf)
print('Recall Regresi Logistik: ', recall_lr)
print('Recall Decision Tree: ', recall_dt)
```

**29. F1**
F1 = 2 x (Precision*Recall)/(Precision+Recall)

```python
from sklearn.metrics import f1_score

# f1: 2 tp / (2 tp + fp + fn)
# Random Forest
f1_rf = f1_score(y_test, y_pred_rf)
# Regresi Logistik
f1_lr = f1_score(y_test, y_pred_lr)
# Decision Tree
f1_dt = f1_score(y_test, y_pred_dt)

print('F1 score Random Forest: ', f1_rf)
print('F1 score Regresi Logistik: ', f1_lr)
print('F1 score Decision Tree: ', f1_dt)
```

**30. ROC**
ROC = TN/FP

Kurva ROC dibuat berdasarkan nilai telah didapatkan pada perhitungan dengan confusion matrix, yaitu antara False Positive Rate dengan True Positive Rate. Dimana:
False Positive Rate (FPR) = FP / (FP + TN)
True Positive Rate (TPR) = TP / (TP + FN)

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

# menghitung fpr dan tpr untuk semua thresholds dari klasifikasi
# Random Forest
probs_rf = model_rf.predict_proba(x_test)
preds_rf = probs_rf[:,1]
fpr_rf, tpr_rf, threshold_rf = roc_curve(y_test, preds_rf)
# Regresi Logistik
probs_lr = model_lr.predict_proba(x_test)
preds_lr = probs_lr[:,1]
fpr_lr, tpr_lr, threshold_lr = roc_curve(y_test, preds_lr)
# Decision Tree
probs_dt = model_dt.predict_proba(x_test)
preds_dt = probs_dt[:,1]
fpr_dt, tpr_dt, threshold_dt = roc_curve(y_test, preds_dt)

# Menghitung Area Under Curve untuk tiap model
# Random Forest
auc_rf = roc_auc_score(y_test, preds_rf)
print('AUC Random Forest: ', auc_rf)
# Regresi Logistik
auc_lr = roc_auc_score(y_test, preds_lr)
print('AUC Regresi Logistik: ', auc_lr)
# Decision Tree
auc_dt = roc_auc_score(y_test, preds_dt)
print('AUC Decision Tree: ', auc_dt)
```

```python
# Menggambarkan plot ROC
# Membuat kanvas
plt.figure(figsize=(10,8))

# plot ROC untuk Random Forest
plt.plot(fpr_rf, tpr_rf,
         label='Kurva ROC Random Forest (luas = {0:0.2f})'
               ''.format(auc_rf),
         color='deeppink', linestyle=':', linewidth=4)

# plot ROC untuk Regresi Logistik
plt.plot(fpr_lr, tpr_lr,
         label='Kurva ROC Regresi Logistik (luas = {0:0.2f})'
               ''.format(auc_lr),
         color='navy', linestyle=':', linewidth=4)

# plot ROC untuk Decision Tree
plt.plot(fpr_dt, tpr_dt,
         label='Kurva ROC Decision Tree (luas = {0:0.2f})'
               ''.format(auc_dt),
         color='orange', linestyle=':', linewidth=4)

# Plot garis diagonal (AUC=0.5)
plt.plot([0, 1], [0, 1], 'k--')

# Set batas dan range sumbu x dan y
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])

#Memberi nama sumbu, judul
plt.ylabel('False Positive Rate')
plt.xlabel('True Positive Rate')
plt.title('Grafik ROC')

# Menambah legenda
plt.legend(loc="lower right")

# Menampilkan plot
plt.show()
```
